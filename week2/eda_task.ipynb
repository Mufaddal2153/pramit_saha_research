```python
import os

for i in range(1, 103):
    os.mkdir(to + "/" + str(i))
```

```python
for i in images_folder:
    for j in os.listdir("102 flower/flowers/valid/" + i):
        shutil.copy("102 flower/flowers/valid/" + i + "/" + j, to + "/" + i + "/" + j)
```

```python
count = 0
for i in os.listdir(to):
    count += len(os.listdir(to + "/" + i))
print(f"Count of images: {count}")
```

```python
import json
label = json.load(open("cat_to_name.json"))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = {}
img_folders = os.listdir(to)
img_folders.sort(key=lambda x: int(x))
for i in img_folders:
    df[label[i]] = len(os.listdir(to + "/" + i))
```

```python
df
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(rotation=-90)
sns.barplot(x=list(df.keys()), y=list(df.values()), palette="viridis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import json

to = "image_data"
label = json.load(open("cat_to_name.json"))

plt.figure(figsize=(20, 20))
for i in range(1, 13):
    plt.subplot(3, 4, i)
    img = mpimg.imread(to + "/" + str(i) + "/" + os.listdir(to + "/" + str(i))[i])
    plt.imshow(img)
    plt.title(label[str(i)])
    plt.axis("off")
```
